### Exercício 1: Similaridade por Cosseno.

Com base na metodologia apresentada na Prática 2, usando o corpus de Machado, Construa vetorizações, binárias, por frequência e por tf-idf. Compare a ordenação dos resultados para uma consulta qualquer, com pelo menos 3 palavras. Para a ordenação dos resultados, use o cosseno entre os vetores da consulta e dos documentos.

```python
from nltk.corpus import machado
from nltk.tokenize import WordPunctTokenizer
from nltk.corpus import stopwords
import string


from collections import defaultdict
from nltk.stem.snowball import PortugueseStemmer
import nltk
import numpy as np
from numpy.linalg import norm
%pylab inline
```

```python
textos = [machado.raw(id) for id in machado.fileids()]
```

```python
#Textos tokenizados
tt = [WordPunctTokenizer().tokenize(t.lower()) for t in textos]
```

```python
voc = set ([])
for t in tt:
    voc = voc.union(set(t))
```

```python
T = nltk.TextCollection(tt)
```

```python
#palavras exemplo
q = "Você camponesa nobre coração todos dias colher lenha bosque"
qt = WordPunctTokenizer().tokenize(q.lower())
```

```python
#definindo a matriz preenchida com zeros
tfm = np.zeros((len(tt),len(qt)))
for j,w in enumerate(qt):
    for i,q in enumerate(tt):
        tfm[i,j] = T.tf(w,q)
        
tfm[:3,:]
```

$$\text{similarity} = \cos(\theta) = {\mathbf{A} \cdot \mathbf{B} \over \|\mathbf{A}\| \|\mathbf{B}\|} = \frac{ \sum\limits_{i=1}^{n}{A_i  B_i} }{ \sqrt{\sum\limits_{i=1}^{n}{A_i^2}}  \sqrt{\sum\limits_{i=1}^{n}{B_i^2}} }$$

Vamos calcular a norma L2 da matriz de tf-idf:
$||A||_F = \sqrt{\sum_{i} |A_{i}|^2}$

```python
MN = np.array([r/norm(r) if norm(r) !=0 else np.zeros(len(r)) for r in tfm])
def ordem(q):
    return [np.dot(q,r) for r in MN]

qv = np.array([T.tf_idf(w,qt) for w in qt])
qv /= norm(qv)
r = ordem(qv)
v=filter(lambda x : x[0]!=0.0, zip(r,range(len(tt))))

resposta = sorted(v, reverse=True) 
resposta
```
